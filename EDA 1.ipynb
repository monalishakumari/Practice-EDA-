```python
# Find null values

# df.isnull()  - inform in form of boolean (true or false)
df.isnull().sum()
```

```python
[features for features in df.columns if df[features].isnull().sum()>0]

# to find null value columns
```

```python
sns.heatmap(df.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

```python
df_country=pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
df.columns
```

```python
# combine dataset
final_df=pd.merge(df,df_country,on='Country Code',how='left')
```

```python
final_df.head(2)
```

```python
# check data types
final_df.dtypes
```

```python
final_df.columns
```

```python
final_df.Country.value_counts()
# India have maximum number of zomato transaction may be happening
```

```python
# All country name shown respect to value count by .index stored in variable country_names
country_names=final_df.Country.value_counts().index 
```

```python
# All values in respect of value count for country stored in variable country_val
country_val=final_df.Country.value_counts().values
```

```python
## Pie Chart.Top 3 country that uses zomato
plt.pie(country_val[:3],labels=country_names[:3],autopct='%1.2f%%')
```

From above pie chart
we can observe that-

Zomato maximum number of records or transaction are from India and then US and UK respectively.

```python
final_df.columns
```

```python
final_df.groupby(['Aggregate rating','Rating color','Rating text'])
```

```python
final_df.groupby(['Aggregate rating','Rating color','Rating text']).size()
```

```python
final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

```python
## Obervation
1. when rating between 4.5-4.9 ---> Excellent
2. when rating between 4.0-4.4 ---> Very Good
3. when rating between 3.5-3.9 ---> Good
4. when rating between 2.5-3.4 ---> Average
5. when rating between 1.8-2.4 ---> Poor
```

```python
ratings.head()
```

```python
import matplotlib

matplotlib.rcParams['figure.figsize']=(15,8)

sns.barplot(x='Aggregate rating',y='Rating Count',data=ratings)
```

```python
sns.barplot(x='Aggregate rating',y='Rating Count',hue='Rating color',data=ratings,palette=['White','Red','Orange','Yellow','Green','Green'])
```

## Obervation

1. Not Rated count is very high
2. Maximum number of rating are between 2.5 to 3.4

```python
##Countplot
sns.countplot(x='Rating color',data=ratings,palette=['White','Red','Orange','Yellow','Green','Green'])

## this is not count , it shows frequency of color 
```
